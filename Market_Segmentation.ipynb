# Developing K-Means Model

```python
pd.set_option("display.max_columns", None)
```

```python
from sklearn.preprocessing import StandardScaler, normalize
```

```python
scaler = StandardScaler()
```

```python
marketdata_scaled = scaler.fit_transform(test_df)
```

```python
marketdata_scaled.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
```

```python
test_df.Income.isnull
```

```python
test_df.isnull().any()
```

```python
test_df['Income'].fillna(0, inplace = True)
```

```python
test_df.isnull().any()
```

```python
marketdata_scaled = scaler.fit_transform(test_df)
```

Elbow Method

```python
scores_1 = []

range_values = range(1, 20)

for i in range_values:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(test_df)
    scores_1.append(kmeans.inertia_) 

plt.plot(scores_1, 'bx-')
plt.title('Finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Scores') 
plt.show()
```

```python
scores_1 = []

range_values = range(1, 25)

for i in range_values:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(marketdata_scaled)
    scores_1.append(kmeans.inertia_) 

plt.plot(scores_1, 'bx-')
plt.title('Finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Scores') 
plt.show()
```

# K-Mean model (K=9)

```python
kmeans = KMeans(9)
kmeans.fit(marketdata_scaled)
labels = kmeans.labels_
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [test_df.columns])
```

# PCA Analysis 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principal_comp = pca.fit_transform(marketdata_scaled)
```

```python
principal_comp
```

```python
pca_df = pd.DataFrame(data = principal_comp, columns =['pca1','pca2'])
pca_df.head()
```

```python
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})], axis = 1)
pca_df.head()
```

```python
plt.figure(figsize=(12,10))
ax = sns.scatterplot(x="pca1", y="pca2", hue = "cluster", data = pca_df, palette =['red','green','blue','pink','yellow','orange','purple', 'black','brown'])
plt.show()
```
